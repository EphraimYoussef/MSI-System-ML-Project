```python
import numpy as np
from sklearn.preprocessing import Normalizer
import joblib
```

## Load features

```python
X_train = np.load("../features/X_train_cnn.npy")
X_val = np.load("../features/X_val_cnn.npy")
X_test = np.load("../features/X_test_cnn.npy")
y_train = np.load("../features/y_train.npy")
y_val = np.load("../features/y_val.npy")
y_test = np.load("../features/y_test.npy")
```

## Original Features

```python
print(f"Original shapes:")
print(f"  Train: {X_train.shape}")
print(f"  Val:   {X_val.shape}")
print(f"  Test:  {X_test.shape}")
```

## Feature Scaling

```python
print("\nScaling features...")
scaler = Normalizer(norm="l2")
X_train_scaled = scaler.fit_transform(X_train)
X_val_scaled = scaler.transform(X_val)
X_test_scaled = scaler.transform(X_test)
```

## Save Features

```python
np.save("../features/X_train_scaled.npy", X_train_scaled)
np.save("../features/X_test_scaled.npy", X_test_scaled)
np.save("../features/X_val_scaled.npy", X_val_scaled)

joblib.dump(scaler, "../models/scaler.pkl")

print(" Feature scaling completed")
```
